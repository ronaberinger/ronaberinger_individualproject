## Data Description of players.csv Dataset

In the **players.csv dataset**, we have: 

- 196 Observations
- 7 variables:
  - experience- A character variable that displays the level of experience a player has, from Beginner -> Regular -> Amateur -> Pro -> Veteran
  - subscribe- A logical variable (true or false) that indicates if a player is subscribed to the game-related newsletter.
  - hashedEmail- A character variable that discloses players' hashed email addresses.
  - played_hours- A double variable that details the number of hours spent playing Minecraft.
  - name- A character variable that reports a player's name.
  - gender- A character variable that states a player's gender.
  - Age- double variable that indicates a player's age. 


The dataset appears to be **tidy**. Each variable has its own column, which are properly named. Each row is an observation for an individual player. Finally, each cell contains only one value

A few issues I see in the data are a couple missing values. Devin and Ahmed are both missing their ages. Additionally, the distribution of the data is a bit skewed. The majority of players have a playing time closest to 0 hours, which could heavily influence statistical analysis. We may need to perform standardization. Furthermore, the data has a majority of players who are college-aged, which could affect the conclusions of the study.

Issues I can't see directly relate to sampling bias. The players in this study include only those who voluntarily joined the UBC server, and may not accurately reflect the entire population of Minecraft players. Moreover, the dataset only records the actions of the players while they are connected to the server, which could limit the predictive power of the data analysis.
